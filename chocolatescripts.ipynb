```python
import pandas as pd
from geopy.geocoders import Nominatim
geolocator = Nominatim()
choco_db = pd.read_csv(r'C:\CS 185c\Final Project\flavors_of_cacao_converted.csv')
locationlist = choco_db['Location'].tolist()
beanlist = choco_db['BroadBeanOrigin'].tolist()
totallist = locationlist + beanlist
totallist2 = set(totallist)


print("Done")
totallist2
countrydict = {}
for x in totallist2:
    a = geolocator.geocode(x)
    print(a)
    countrydict[x] = a
```

```python
from geopy import distance
countrydict
locationlist = choco_db['Location'].tolist()
beanlist = choco_db['BroadBeanOrigin'].tolist()
loclatlong = []
for location in locationlist:
    loc = countrydict[location]
    try:
        loclatlong.append((loc.latitude, loc.longitude))
    except AttributeError:
        loclatlong.append("NULL")
    
beanlatlong = []
for bean in beanlist:
    b = countrydict[bean]
    try:
        beanlatlong.append((b.latitude, b.longitude))
    except AttributeError:
        beanlatlong.append("NULL")

dist = []
for i in range(len(locationlist)):
    try:
        dist.append(distance.distance(loclatlong[i], beanlatlong[i]).miles)
    except ValueError:
        dist.append('NULL')
choco_db['loclatlong'] = loclatlong
choco_db['beanlatlong'] = beanlatlong
choco_db['dist'] = dist
choco_db.head(50)

choco_db2 = choco_db[choco_db['beanlatlong'] != 'NULL']
choco_db3 = choco_db2[choco_db2['loclatlong'] != 'NULL']
choco_db3.to_csv('finaldist.csv')


```
